--- Part Two ---

Time to check the rest of the slopes - you need to minimize the probability of a sudden arboreal stop, after all.

Determine the number of trees you would encounter if, for each of the following slopes, you start at the top-left corner and traverse the map all the way to the bottom:

- Right 1, down 1.
- Right 3, down 1. (This is the slope you already checked.)
- Right 5, down 1.
- Right 7, down 1.
- Right 1, down 2.

In the above example, these slopes would find 2, 7, 3, 4, and 2 tree(s) respectively; multiplied together, these produce the answer 336.

What do you get if you multiply together the number of trees encountered on each of the listed slopes?

````
# Find the number of trees encountered
def num_trees_encountered(go_right, go_down, map):
  grid_size = [len(map[0]), len(map)]

  index = [0, 0]
  tree_count = 0
  while index[1] < (grid_size[1] - go_down):
    index[0] = (index[0] + go_right) % grid_size[0]
    index[1] += go_down 
    if map[index[1]][index[0]] == '#':
      tree_count += 1
  return tree_count
````

````
# Find the product of trees enountered for the following slopes

tree_count = (num_trees_encountered(1, 1, map) * 
              num_trees_encountered(3, 1, map) * 
              num_trees_encountered(5, 1, map) * 
              num_trees_encountered(7, 1, map) * 
              num_trees_encountered(1, 2, map))

print('Product of number of trees encountered:', tree_count)
````
